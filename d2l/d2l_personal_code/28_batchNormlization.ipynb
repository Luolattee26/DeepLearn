前面我们讲的conv、max pool这些思想，其实在上个世纪就已经变成神经网络里面的层了（比如LeNet）

但是batch normalization这个思想虽然也挺老的，实际上用到深度神经网络中，变成一个层，也就是最近几年的事情
* 在我们的深度网络训练的时候，前向传播其实没啥，但是反向传播由于存在n个梯度相乘，并且梯度通常比较小（当然也会有梯度爆炸），所以深层网络梯度比较正常，但是浅层网络的梯度容易变得太小
* 因为深层网络相当于计算求导的时候相乘的次数还比较小，但是浅层的要乘好多次
* 这导致一个情况就是，深层的参数更新和浅层的并不同步，具体说来：
    * 由于深层的梯度比较大，所以容易收敛
    * 但是浅层梯度小，所以收敛慢
    * 浅层我们直观的可以理解，是对数据的特征做了基础的抽取，而深层网络则是在浅层学习到的特征上再进行学习，拟合最终的函数
    * 这也就意味着，浅层的收敛慢，导致深层收敛快也是没用的，因为你浅层的参数还在变化呢

batch normalization想要解决的就是上面提到的这个问题，既然深层浅层的梯度肯定有差别，那我能不能尽量让浅层还在训练的时候，深层的参数可以变化不大呢（这样就算有变化，深层再学习比较小的变化收敛起来也很快）
* 还记得在前面数值稳定性的时候，我们提过如果数据的均值和方差能尽量保持不变，维持在同一个分布上，那么整个网络会比较稳定（也就是大家学的东西都差不多，自然不用担心哪里崩掉了）
* 现在batch normalization其实也是这个思路，因为我们是按照每个batch扔进去训练，所以我们想让这一个batch在整个网络中，均值和方差始终维持差不多的一个分布
* 放到具体的数据上，也就是每次数据进来，我先将它变成一个均值0方差1的标准分布，然后我再对其进行一个线性变换（用学出来的另一个方差和均值），这样我就能控制我神经网络中，这一个批次数据的分布情况了，并且由于是可学习的参数，所以这也不用我们自己去想这个参数咋设置
* 其实也就是，总的来说既然我想要让我的这个batch的数据尽量稳定，那我就先让他变成最为稳定的01分布，但是这样可能太暴力了，所以我就再学习两个新的系数，我允许有一定的偏差（因为这两个系数的大小是可以控制的），这样我一个batch的数据在神经网络中传播的时候并不会变化很大，从而让网络更加稳定

batch normalization作用在具体的层上的时候：
1. 全连接和卷积的输出上，在relu之前（因为看公式可以发现，bn其实是一个线性变换，如果先做relu再做bn，那不就是把非线性又改回线性，非常老残）
2. 全连接和卷积的输入上
    1. 对全连接层，作用在特征维度上（因为注意，批次这个概念是针对多个样本来说的）
    2. 对卷积层，作用在通道维度上

bn层在论文里面和后续研究上是有显著差别的：
1. 最开始论文里面提出来的，其实认为是bn层减少了内部协变量的转移（也就是不同时间点上，世界状态不一样，比如疫情前后，所以模型想要尽量减少这种协变量的转移或者说变化，这玩意意会就行）
2. 后面大家论文里面又觉得，bn层实际上是往数据里面加了噪音从而提升模型鲁棒性（dropout既视感）
3. 这个噪音咋来的，回到公式我们在转换成01标准分布的时候，我们是用了这一个batch的均值和方差来算的，但是还记得吗，我们的优化方法叫随机小批量梯度下降。随机小批量，这样的采样方式就导致了我们拿到的这个batch的均值和方差肯定那是不准确含有噪音的，所以我们作bn函数，也引入了这个噪音
4. 不过上面这个理解，也不一定是正确的，唯一正确的就是bn层真的有用......
3. 所以bn和dropout没必要混用，两个在干一个事
4. bn层最大的作用是加快收敛，但是对精度没啥帮助（你模型可以搞的深一点）

下面我们来实现一下bn层

```python
import torch
from torch import nn
from d2l import torch as d2l


def batch_norm(X, gamma, beta, moving_mean, moving_var, eps, momentum):
    # 通过is_grad_enabled来判断当前模式是训练模式还是预测模式
    if not torch.is_grad_enabled():
        # 如果是在预测模式下，直接使用传入的移动平均所得的均值和方差
        X_hat = (X - moving_mean) / torch.sqrt(moving_var + eps)
    else:
        assert len(X.shape) in (2, 4)
        if len(X.shape) == 2:
            # 使用全连接层的情况，计算特征维上的均值和方差
            mean = X.mean(dim=0)
            var = ((X - mean) ** 2).mean(dim=0)
        else:
            # 使用二维卷积层的情况，计算通道维上（axis=1）的均值和方差。
            # 这里我们需要保持X的形状以便后面可以做广播运算
            mean = X.mean(dim=(0, 2, 3), keepdim=True)
            var = ((X - mean) ** 2).mean(dim=(0, 2, 3), keepdim=True)
        # 训练模式下，用当前的均值和方差做标准化
        X_hat = (X - mean) / torch.sqrt(var + eps)
        # 更新移动平均的均值和方差
        moving_mean = momentum * moving_mean + (1.0 - momentum) * mean
        moving_var = momentum * moving_var + (1.0 - momentum) * var
    Y = gamma * X_hat + beta  # 缩放和移位
    return Y, moving_mean.data, moving_var.data # 这里其实是因为我们后面定义bn层，这里会是tensor
```

上面就是手动实现一个比较简单的bn层的代码（因为只支持全连接层、2d卷积），有些难看懂，这里讲解一下：
1. 首先，moving_mean, moving_var这两玩意是啥呢？这两玩意其实是为interface的时候准备的，也就是预测的时候。因为我们算均值方差是在batch水平上，这是训练，预测的时候可能就一张图片扔进来，让你预测，这时候你咋算batch的均值方差？所以moving_mean, moving_var表示的是一个全局均值方差
2. eps这玩意就是防止有时候方差0，加上一点点，不然除0整个网络的计算可能就崩掉了
3. 下面发现我们用momentum这玩意去更新了全局的mean var，这个算法是信号处理那边来的，momentum一般是0.9或者0.1，这样可以通过你现在算出来的batch样本均值方差不断逼近真实值
4. 然后我们用学习出来的gamma值，对数据进行batch normalization

```python
class BatchNorm(nn.Module):
    # num_features：完全连接层的输出数量或卷积层的输出通道数。
    # num_dims：2表示完全连接层，4表示卷积层
    def __init__(self, num_features, num_dims):
        super().__init__()
        if num_dims == 2:
            shape = (1, num_features)
        else:
            shape = (1, num_features, 1, 1)
        # 参与求梯度和迭代的拉伸和偏移参数，分别初始化成1和0
        self.gamma = nn.Parameter(torch.ones(shape))
        # gamma是拟合出来的方差，不能是0
        self.beta = nn.Parameter(torch.zeros(shape))
        # 非模型参数的变量初始化为0和1
        self.moving_mean = torch.zeros(shape) # 这两玩意反正不迭代，所以用tensor就可以了
        self.moving_var = torch.ones(shape)

    def forward(self, X):
        # 如果X不在内存上，将moving_mean和moving_var
        # 复制到X所在显存上
        if self.moving_mean.device != X.device:
            self.moving_mean = self.moving_mean.to(X.device)
            self.moving_var = self.moving_var.to(X.device)
        # 保存更新过的moving_mean和moving_var
        Y, self.moving_mean, self.moving_var = batch_norm(
            X, self.gamma, self.beta, self.moving_mean,
            self.moving_var, eps=1e-5, momentum=0.9) # eps这玩意每个框架都不太一样 # 0.9是正常取值
        return Y
```

```python
net = nn.Sequential(
    nn.Conv2d(1, 6, kernel_size=5), BatchNorm(6, num_dims=4), nn.Sigmoid(),
    nn.AvgPool2d(kernel_size=2, stride=2),
    nn.Conv2d(6, 16, kernel_size=5), BatchNorm(16, num_dims=4), nn.Sigmoid(),
    nn.AvgPool2d(kernel_size=2, stride=2), nn.Flatten(),
    nn.Linear(16*4*4, 120), BatchNorm(120, num_dims=2), nn.Sigmoid(),
    nn.Linear(120, 84), BatchNorm(84, num_dims=2), nn.Sigmoid(),
    nn.Linear(84, 10)) # 用起来其实和别的layer没啥区别
```

```python
lr, num_epochs, batch_size = 1.0, 10, 256
train_iter, test_iter = d2l.load_data_fashion_mnist(batch_size)
d2l.train_ch6(net, train_iter, test_iter, num_epochs, lr, d2l.try_gpu())
```
